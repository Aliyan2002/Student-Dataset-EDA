# Bi-Variate EDA

```python
# Using correlation matrix to find out the relation between features
cor=df.corr()
sns.heatmap(cor,annot=True)
```

```python
1. Height is Highly Correlated with Length, Diameter, All Weights.
2. Length is Highly Correlated with Diameter, Height, All Weights.
3. Diameter is Highly Correlated with Length, Height, All Weights.
4. All weights are Highly Correlated with Length, Height, Diameter.
5. Our Target Variables i.e. Rings and Age have higher correlation with Length, Height, Diameter, Shell Weight.
```
